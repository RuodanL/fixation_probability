```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mathtext as mathtext
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

mathtext.FontConstantsBase.sup1 = 0.25
plt.rcParams["font.family"] = "sans-serif"
plt.rcParams["font.sans-serif"] = "Arial"
plt.rcParams['mathtext.fontset'] = "custom"
plt.rcParams["mathtext.rm"] = "Arial"
plt.rcParams['mathtext.it'] = 'Arial:italic'

rows, cols = 2, 2
fig, ax = plt.subplots(rows, cols, sharey = 'row', figsize=(10,8))

r=np.arange(0.01,3.00,0.0001)
fix_moran=(1-1/r)/(1-1/(r**4))    #fixation probability for Moran process with N=4
fix_star=(3*r**3+5*r**2)/(12*r**3+56*r**2+72*r+36)+(9*r**2)/(12*r**2+20*r+12)
                                  #fixation probability for star 3-uniform hypergraph with N=4
fix_star_omp=np.loadtxt('omp-star_N4_numerical.txt')
                                  #fixation probability for one-mode projection of star hypergraph with N=4

#ax[0,0].set_title('ER, $q^*=0.1$, slow', fontsize=14)
ax[0,0].tick_params(labelsize=15)
ax[0,0].set_xlim([0.0, 3.0])
#ax[0,0].set_xticklabels(['0','5','10','15','20'])
ax[0,0].set_ylim([-0.02, 0.7])
#ax[0,0].set_yticks([0, 0.5, 1])
#ax[0,0].set_yticklabels(['0', '0.5', '1'])
ax[0,0].plot(r, fix_moran, linewidth=1.5, label='Moran', color='black')
ax[0,0].plot(r, fix_star, linewidth=1.5, label='star hypergraph')
ax[0,0].plot(r, fix_star_omp, linewidth=1.5, label='one-mode projection')
#legend_font = {'family' : 'Arial', 'weight' : 'normal', 'size': 12}
#ax[0,0].legend(prop=legend_font)
ax[0,0].legend(fontsize=12, frameon=False)
fig.text(0.1, 0.91, '(a)', fontsize=18)
fig.text(0.27, 0.905, '$N=4$', fontsize=16)

r=np.arange(0.01,3.00,0.0001)
fix_moran=(1-1/r)/(1-1/(r**5))      #fixation probability for Moran process with N=5
fix_star=(72*r**5+202*r**4+145*r**3+6*r**2)/(360*r**5+2450*r**4+5125*r**3+5350*r**2+4400*r+1440)+(288*r**4+376*r**3+16*r**2)/(360*r**4+1010*r**3+1085*r**2+1010*r+360)
                                    #fixation probability for star 3-uniform hypergraph with N=5
fix_star_omp=np.loadtxt('omp-star_N5_numerical.txt')
                                    #fixation probability for one-mode projection of star hypergraph with N=5

#ax[0,2].set_title('ER, $q^*=0.5$, slow', fontsize=14)
ax[0,1].tick_params(labelsize=15)
ax[0,1].set_xlim([0.0, 3.0])
#ax[0,1].set_xticklabels(['0', '1', '2'])
#ax[0,1].set_ylim([0.0, 1.0])
ax[0,1].plot(r, fix_moran, linewidth=1.5, label='Moran', color='black')
ax[0,1].plot(r, fix_star, linewidth=1.5, label='3-star')
ax[0,1].plot(r, fix_star_omp, linewidth=1.5, label='one mode proj')
fig.text(0.513, 0.91, '(b)', fontsize=18)
fig.text(0.69, 0.905, '$N=5$', fontsize=16)

N=20
r=np.arange(0.01,3.00,0.0001)
fix_moran=(1-1/r)/(1-1/(r**N))
fix_star=np.loadtxt('3-star_N20_numerical.txt')
fix_star_omp=np.loadtxt('omp-star_N20_numerical.txt')

#ax[2,0].set_title('Collabo., $q^*=0.1$, slow', fontsize=14)
ax[1,0].tick_params(labelsize=15)
ax[1,0].set_xlim([0.0, 3.0])
#ax[1,0].set_xticklabels(['0', '10', '20', '30', '40'])
ax[1,0].set_ylim([-0.02, 0.7])
#ax[1,0].set_yticks([0, 0.5, 1])
#ax[1,0].set_yticklabels(['0', '0.5', '1'])
ax[1,0].plot(r, fix_moran, linewidth=1.5, label='Moran', color='black')
ax[1,0].plot(r, fix_star, linewidth=1.5, label='3-star')
ax[1,0].plot(r, fix_star_omp, linewidth=1.5, label='one mode proj')
fig.text(0.1, 0.46, '(c)', fontsize=18)
fig.text(0.27, 0.455, '$N=20$', fontsize=16)

#inset_axes = inset_axes(ax[2,0], width="50%", height=0.8, loc=1)
inset_axes = fig.add_axes([0.18, 0.3, 0.1, 0.1])
inset_axes.set_xlim([0.9, 1.0])
inset_axes.set_xticks([0.9, 0.95, 1.0])
inset_axes.set_xticklabels(['0.9', '0.95', '1.0'])
inset_axes.set_ylim([0.01, 0.05])
inset_axes.set_yticks([0.01, 0.03, 0.05])
inset_axes.set_yticklabels(['0.01', '0.03', '0.05'])
inset_axes.set_xlabel('$r$', fontsize=12)
inset_axes.plot(r, fix_moran, linewidth=1.5, label='Moran', color='black')
inset_axes.plot(r, fix_star, linewidth=1.5, label='3-star')
inset_axes.plot(r, fix_star_omp, linewidth=1.5, label='one mode proj')

N=200
r=np.arange(0.01,3.00,0.0001)
fix_moran=(1-1/r)/(1-1/(r**N))
fix_star=np.loadtxt('3-star_N200_numerical.txt')
fix_star_omp=np.loadtxt('omp-star_N200_numerical.txt')

#ax[2,2].set_title('Collabo., $q^*=0.5$, slow', fontsize=14)
ax[1,1].tick_params(labelsize=15)
ax[1,1].set_xlim([0.0, 3.0])
#ax[1,1].set_xticklabels(['0', '2', '4', '6'])
#ax[1,1].set_ylim([0.0, 1.0])
ax[1,1].plot(r, fix_moran, linewidth=1.5, label='Moran', color='black')
ax[1,1].plot(r, fix_star, linewidth=1.5, label='3-star')
ax[1,1].plot(r, fix_star_omp, linewidth=1.5, label='one mode proj')
fig.text(0.513, 0.46, '(d)', fontsize=18)
fig.text(0.68, 0.455, '$N=200$', fontsize=16)

#inset_axes = inset_axes(ax[2,0], width="50%", height=0.8, loc=1)
inset_axes = fig.add_axes([0.61, 0.3, 0.1, 0.1])
inset_axes.set_xlim([0.99, 1.0])
inset_axes.set_xticks([0.99, 0.995, 1.0])
inset_axes.set_xticklabels(['0.99', '0.995', '1.0'])
inset_axes.set_ylim([0.001, 0.005])
inset_axes.set_yticks([0.001, 0.003, 0.005])
inset_axes.set_yticklabels(['0.001', '0.003', '0.005'])
inset_axes.set_xlabel('$r$', fontsize=12)
inset_axes.plot(r, fix_moran, linewidth=1.5, label='Moran', color='black')
inset_axes.plot(r, fix_star, linewidth=1.5, label='3-star')
inset_axes.plot(r, fix_star_omp, linewidth=1.5, label='one mode proj')

fig.text(0.5, 0.06, '$r$', ha='center', va='center', fontsize=28)
fig.text(0.03, 0.5, 'Fixation probability', ha='center', va='center', rotation='vertical', fontsize=28)
plt.subplots_adjust(hspace=0.5)
#plt.tight_layout()
plt.savefig('star-and-its-omp.pdf', bbox_inches="tight")
plt.show()
```
